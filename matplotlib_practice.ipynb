```python
# Create the bar chart

plt.bar(x_axis, y_axis, color="green", label="Boston")
plt.ylim(0,50)
plt.xlabel('Months')
plt.ylabel('Fares($)')
plt.title('PyBer Fare by Month')
#plt.set_axisbelow(True)
plt.legend()
```

# Now use the object-oriented approach

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="indigo", label="Chicago")
ax.set_ylim(ax.get_ylim()[::-1])
ax.set_axisbelow(True)
ax.set_ylabel('Months')
ax.set_xlabel('Fares($)')
ax.set_title('PyBer Fares by Month')
#ax.gca().invert_yaxis()
ax.grid()
ax.legend()
```

```python
plt.plot(y_axis, x_axis, 'o', color='red', label='Chicago')
#plt.ylim(-1,12)
plt.xlim(0,50)
plt.gca().invert_yaxis()
plt.ylabel('Date')
plt.xlabel('Fare($)')
plt.title('PyBer Fare by Month')
plt.grid()
plt.legend()
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# The following is the list comprehension technique
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis], 
           label="Boston", color='skyblue', alpha=0.8,
          linewidth=2, edgecolor='k')
ax.legend()
ax.grid()
ax.set_ylim(ax.get_ylim()[::-1])
ax.set_xlim(0,50)
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.set_axisbelow(True)
```

# Same as above but using the object-oriented approach

```python
colors = ["slateblue", 
          "magenta", 
          "lightblue", 
          "green", 
          "yellowgreen", 
          "greenyellow", 
          "yellow", 
          "orange", 
          "gold", 
          "indianred", 
          "tomato", 
          "mistyrose"]
fig, ax = plt.subplots(figsize=(8,8))
explode_values = [0,0,0.2,0,0,0,0.2,0,0,0,0,0]
ax.pie(y_axis, 
       labels=x_axis,
       autopct='%.1f%%',
       shadow=True,
       colors=colors,
       explode=explode_values,
       startangle=-95)
ax.set_ylim(ax.get_ylim()[::-1])
plt.show()
```
